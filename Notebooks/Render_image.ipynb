```python
from PIL import Image
```

```python
flower  = Image.open('./Data/flower.jpg')
```

```python
print(flower.size,flower.format,flower.mode,sep=",")
```

```python
crop_butterfly = flower.crop((55,100,120,180))
```

```python
#blur the cropped image and paste in original
from PIL import ImageFilter
sq_box = (55,100,120,180)
crop_butterfly = crop_butterfly.filter(ImageFilter.BLUR)
```

```python
crop_butterfly
```

```python
flower.paste(crop_butterfly,sq_box)
```
